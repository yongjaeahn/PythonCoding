방정식 풀기

```python
import sympy as sp

x = sp.Symbol('x')
fx1 = 2*x - 3
print('방정식 : f(x)=',fx1)
root = sp.solve(fx1)
print('방정식의 근 :', root)
y = fx1.subs({x:root[0]})
print('근을 사용한 방정식의 계산 결과 :', y)
```

```python
import sympy as sp

x = sp.Symbol('x')
fx3 = 3*x**3 - 2*x**2 + 4
print('방정식 : f(x)=',fx3)
roots = sp.solve(fx3)
print('방정식의 근 :', roots)
for root in roots:
    print('방정식의 근 :', root)
```

```python
from math import sqrt

for root in roots:
    print('방정식의 근 :', eval(str(root).replace('I','1j')))
    
print('근의 객체유형 :', type(roots[0]))
```

```python
import sympy as sp

x = sp.Symbol('x')
a = sp.Symbol('a')
b = sp.Symbol('b')

fx2 = a*x**2 + b*x + 1
print('방정식 : f(x)=',fx2)
roots = sp.solve(fx2,'x')
print('미지수 x에 대한 방정식의 근 :', roots)
roots = sp.solve(fx2,'a')
print('미지수 a에 대한 방정식의 근 :', roots)
roots = sp.solve(fx2,'b')
print('미지수 b에 대한 방정식의 근 :', roots)
```

```python
import sympy as sp

x = sp.Symbol('x')
fx2 = x**2 + 2*x + (3-3j)
print('방정식 : f(x)=',fx2)
roots = sp.solve(fx2)
print('방정식의 근 :', roots)

for root in roots:
    y = fx2.subs({x:root})
    print('근을 사용한 방정식의 계산 결과 :', y)
    y = eval(str(y).replace('I','1j'))
    print('계산결과가 0과 같은지 여부 :',math.isclose(abs(y),0,abs_tol=1e-14))
```

```python
import sympy as sp

x = sp.Symbol('x')
y = sp.Symbol('y')
z = sp.Symbol('z')

f1 = sp.Eq(4*x - 2*y + 3*z,9)
f2 = sp.Eq(3*x + y + 5*z,8)
#f3 = sp.Eq(-2*x + 2*y - 2*z,5)
f3 = -2*x + 2*y - 2*z - 5 

solutions = sp.solve((f1,f2,f3))
print('연립방정식의 해 :', solutions)
for key, value in solutions.items():
    print('연립방정식의 해 :', key, eval(str(value)))

print('심볼의 객체유형 :', type(x))
print('해의 객체유형 :', type(solutions[x]))
```

```python
print('x의 해를 사용한     f1 방정식의 계산 결과 :',f1.subs({x:solutions[x]}))
print('x와 y의 해를 사용한 f1 방정식의 계산 결과 :',f1.subs({x:solutions[x],y:solutions[y]}))

print('해를 모두 사용한 f1 방정식의 계산 결과 :',f1.subs({x:solutions[x],y:solutions[y],z:solutions[z]}))
print('해를 모두 사용한 f2 방정식의 계산 결과 :',f2.subs({x:solutions[x],y:solutions[y],z:solutions[z]}))
print('해를 모두 사용한 f3 방정식의 계산 결과 :',f3.subs({x:solutions[x],y:solutions[y],z:solutions[z]}))
```

```python
import sympy as sp

x = sp.Symbol('x')
fx1 = x + x
print('x + x =', fx1)
fx2 = 2*x**2 + 3*x**2 + 3 - 4 
print('2*x**2 + 3*x**2 + 3 – 4 =', fx2)
```

```python
fx3 = fx1 + fx2
print('fx1 + fx2 =', fx3)
fx4 = fx1 - fx2
print('fx1 - fx2 =', fx4)
print('x**2 * x**2 =', x**2 * x**2)
print('x**4 / x**2 =', x**4 / x**2)
print('fx1**2', fx1**2)
print('3log(x) - exp(x) =',3*sp.log(x) - sp.exp(x))
```

```python
import sympy as sp

x = sp.Symbol('x')
fx2 = x**2 + 1

print(fx2.coeff(x,0))
print(fx2.coeff(x,1))
print(fx2.coeff(x,2))
print(fx2.coeff(x,3))
print(fx2.as_coefficients_dict())
print('x**2의 계수 :', fx2.as_coefficients_dict()[x**2])
```

```python
import sympy as sp

x = sp.Symbol('x')
y = sp.Symbol('y')
z = sp.Symbol('z')

f1 = sp.Eq(4*x - 2*y + 3*z,9)
f3 = -2*x + 2*y - 2*z - 5

print(f3.coeff(x,0))
print(f3.coeff(x,1))
print(f3.coeff(y))
print(f3.coeff(z))
print(f3.as_coefficients_dict())

print(f1.coeff(x))
print(f1.as_coefficients_dict())
```
